```python

import main
import pandas as pd
import os
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.inspection import PartialDependenceDisplay
import xgboost as xgb
from pdpbox import  pdp
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
# data loading
data_pd=pd.read_csv('dataset.csv')

data_clean = data_pd[data_pd['mean_angle'] != 0]
data_clean = data_clean[data_clean['gsa'] < 1e6]
data_clean = data_clean[data_clean['mean_angle'] < 50]
ty1_data=data_clean[data_clean['cca_d'] ==0]
ty2_data=data_clean[data_clean['cca_d'] > 0]


```

```python
ty1_features=['gsa','cve','mean_angle','edges']
ty1_target='cve_h'
ty2_features=['gsa','rel_cve','cca','rel_cveh','mean_angle','edges']
ty2_target='cca_d'
# training-test splitting
ty1_xtrain, ty1_xtest, ty1_ytrain, ty1_ytest \
    = train_test_split(ty1_data[ty1_features].to_numpy(),
                       ty1_data[ty1_target].to_numpy(), test_size=0.2, random_state=42)
ty2_xtrain, ty2_xtest, ty2_ytrain, ty2_ytest \
    = train_test_split(ty2_data[ty2_features].to_numpy(),
                       ty2_data[ty2_target].to_numpy(), test_size=0.2, random_state=42)
```

```python
ty1_x_train, ty1_xval, ty1_y_train, ty1_yval=train_test_split(
    ty1_xtrain,ty1_ytrain, test_size=0.1, random_state=42)
xgb_model_es_1 = main.XGBoostModel(500,4,0.05,42,1,1)
xgb_model_es_1.train_es(ty1_x_train, ty1_y_train,X_val=ty1_xval, y_val=ty1_yval, 
                        early_stopping_rounds=10)

# Evaluate the model
xgb_model_es_1.evaluate(ty1_xtest, ty1_ytest)
'''
xgb_model_2 = main.XGBoostModel(500,5,0.05,42)
xgb_model_2.train(ty2_xtrain, ty2_ytrain)

# Evaluate the model
xgb_model_2.evaluate(ty2_xtest, ty2_ytest)
'''
```
